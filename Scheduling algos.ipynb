```python
'''
you can add many more process as you want in this format given in input.txt
'''
def prLightPurple(skk):
    print("\033[94m {}\033[00m" .format(skk))
    
def prCyan(skk): 
    print("\033[96m {}\033[00m" .format(skk))

def prRed(skk): 
    print("\033[91m{}\033[00m" .format(skk))

def TT(l):
    sum=0
    x=l[len(l)-1][0]
    for i in range(len(l)):
        sum+=l[i][2]
    tpt=x/sum
    prRed("\nOverall throughput = " + str(tpt))
    
def RT(gc,l):
    g=[]
    h=[]
    for i in range(len(gc)):
        if gc[i] not in g:
            g.append(gc[i])
            h.append(i)
    sum=0
    for i in range(len(l)):
        sum+=(h[i]-l[i][1])
    prLightPurple("Average Response time ="+str(sum/len(l)))
    
def TAW(gc,l):
    gc.reverse()
    g=[]
    h=[]
    
    for i in range(len(gc)):
        if gc[i] not in g:
            g.append(gc[i])
            h.append(len(gc)-i)
    s=[]
    for i in range(len(g)):
        s.append([g[i],h[i]])
    s = sorted(s,key = lambda process: process[0])
    
    tt=[]
    ttsum=0
    for i in range(len(s)):
        tt.append(s[i][1]-l[i][1])
        ttsum+=s[i][1]-l[i][1] 
    prLightPurple ("Average TurnAround time ="+str(ttsum/len(l)))
    
    wt =[]
    wtsum=0
    for i in range(len(s)):
        wtsum+=tt[i]-l[i][2]
    prLightPurple("Average Waiting time ="+str(wtsum/len(l)))


def fcfs(l):
    sort_list = sorted(l,key = lambda process: process[1])
    gchart=[]
    print("Gantt chart is : ",end="")
    for i in sort_list:
        print(str(i[0])*i[2],end="")
        for j in range(i[2]):
            gchart.append(i[0])       
    TT(l)
    RT(gchart , l)
    TAW(gchart , l)
    
def sjf(l):
    t = 0
    p = []
    f=[]
    gchart=[]
    print("Gantt chart is : ",end="")
    for i in range(len(l)):
        s = []
        for j in range(len(l)):
            if l[j][1] <= t:
                if l[j][0] not in p:
                    s.append(l[j])
        s = sorted(s,key = lambda process: process[2])
        f.append(s[0][0])
        for j in range(s[0][2]):
            gchart.append(s[0][0]) 
        print(str(s[0][0])*s[0][2],end="")
        p.append(s[0][0])
        t+=s[0][2]
    TT(l)
    RT(gchart , l)
    TAW(gchart , l)
    
    
def srtf(l):
    remaining_time = []
    gchart=[]
    total=0
    print("Gantt chart is : ",end="")
    for i in range(len(l)):
        total+=l[i][2]
    count=0
    d={"":10000000000}
    while count<=total:
        for i in l:
            if i[1] == count:
                d[i[0]] = i[2]
        y=min(d, key=d.get)
        count+=1
        gchart.append(y)
        print(y,end="")
        d[y]=d[y]-1
        if d[y]==0:
            d.pop(y)
            
    gchart.pop()
    TT(l)
    RT(gchart , l)
    TAW(gchart , l)
    
def rr(li,info):
    l = li
    gchart=[]
    qt = int(input("Enter the qauntam time : ")) 
    d = {}
    x = 0
    count = 0
    sum1 = max(l,key = lambda parameter: parameter[1])[1]
    p = []
    for i in l:
        d[i[0]] = i[1]
    for i in range(len(l)):
        if l[i][2]%qt == 0:
            x += int(l[i][2]/qt)
        else:
            x += int(l[i][2]/qt) + 1
    print("Gantt chart is : ",end="")
    for i in range(x):
        s = []
        for j in range(len(l)):
            if l[j][1] <= count:
                if l[j] not in p and l[j][0] != -1:
                    p.append(l[j])
        for k in range(len(p)):
            s.append([p[k][0],count-d[p[k][0]]])
        s = sorted(s,key = lambda item: item[1],reverse = True)
        y = s[0][0]
            
        if l[y-1][2] >= qt:
            print(str(y)*qt,end="")
            count += qt
            l[y-1][2] -= qt
            for i in range(qt):
                gchart.append(l[y-1][0])
        else:
            print(str(y)*l[y-1][2],end="")
            count += l[y-1][2]
            for i in range(l[y-1][2]):
                gchart.append(l[y-1][0])
            l[y-1][2] -= l[y-1][2]
        d[y] = count
        if l[y-1][2] == 0:
            p.remove(l[y-1])
            l[y-1]=[-1,-1,-1]
    TT(info)
    RT(gchart , info)
    TAW(gchart , info)
    

   
os = open("input.txt","r")

info = []
l = os.readlines()

count = 0
for i in l:
    if count==0:
        count+=1
        continue
    else:
        info.append(i.split(" "))

temp = []
for i in range(len(info)): 
    map_object = map(int, info[i])
    new_info = list(map_object)
    temp.append(tuple(new_info))
info = tuple(temp)

while True:
    li  = []
    for i in info:
        li.append(list(i))
    print("\n\n---------------------------")
    prCyan("1.First come First serves")
    prCyan("2.Shortest Job First")
    prCyan("3.Shortest Remainnig Time First")
    prCyan("4.Round Robin")
    prCyan("5.Exit")
    print("---------------------------")
    ch = int(input("Enter your choice: "))
    if ch == 1:
        fcfs(li)
    elif ch == 2:
        sjf(li)
    elif ch == 3:
        srtf(li)
    elif ch == 4:
        rr(li,info)
        #tpt(info)
    elif ch == 5:
        print("Exited\n")
        break
    else:
        print("wrong choice\n")
```
